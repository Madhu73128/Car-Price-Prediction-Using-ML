```python
#importing necessary LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set()
```

```python
#importing DATASET
car_data=pd.read_csv('Car_Raw_Dataset.csv')
cd=car_data#for calculating deleted rows
car_data
```

```python
#Generating SUMMARY STATISTICS for the car data
car_data.describe()
```

```python
#Generating SUMMARY STATISTICS for ALL COLUMNS, including CATEGORICAL DATA
car_data.describe(include='all')
```

```python
#Dropping the 'Model' column from car_data
car_data.drop('Model',axis=1,inplace=True)
print('Column Model of car_data is dropped')

#Generating SUMMARY STATISTICS for ALL COLUMNS, including CATEGORICAL DATA
car_data.describe(include='all')
```

```python
#Checking for the total number of missing values in each column of car_data
car_data.isnull().sum()
```

```python
#Dropping rows with missing or null (NaN) values from car_data
car_data.dropna(axis=0,inplace=True)
print("Missing or Null(Nan) Values are dropped")

#Verifying that there are no remaining missing values in each column of car_data
car_data.isnull().sum()
```

```python
# Displaying the current state of the car_data DataFrame
car_data
```

```python
#Generating SUMMARY STATISTICS for ALL COLUMNS, including CATEGORICAL DATA
car_data.describe(include='all')
```

```python
#Setting the figure size for the plot
plt.figure(figsize=(8,6))

#Plotting the distribution of car prices
sns.distplot(car_data['Price'])

#Adding a title to the plot
plt.title('Car price Distribution')
```

```python
#Calculating the 99th percentile value for the Price column to identify potential outliers
price_outlier=car_data['Price'].quantile(q=0.99)

#Displaying the 99th percentile value for the Price column
price_outlier
```

```python
price_modified_car_data=car_data[car_data['Price']<price_outlier]

price_modified_car_data.describe(include='all')
#modified_car_data
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.distplot(car_data['Price'])
plt.title("Before Removing Price Outliers")

plt.subplot(1,2,2)
sns.distplot(price_modified_car_data['Price'])
plt.xlim(0,129222)#(600,129222)
plt.title('After Removing Price Outliers')
```

```python
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
sns.distplot(car_data['Mileage'])
plt.title('Car Mileage Distribution')

plt.subplot(1,3,2)
sns.distplot(car_data['EngineV'])
plt.title('Car Engine Volume Distribution')

plt.subplot(1,3,3)
sns.distplot(car_data['Year'])
plt.title('Car Year Distribution')
```

```python
mileage_outlier=car_data['Mileage'].quantile(q=0.99)
mileage_outlier
```

```python
enginev_outlier=car_data['EngineV'].quantile(q=0.99)
enginev_outlier
```

```python
year_outlier=car_data['Year'].quantile(q=0.01)
year_outlier
```

```python
mileage_modified_car_data=price_modified_car_data[price_modified_car_data['Mileage']<mileage_outlier]
enginev_modified_car_data=mileage_modified_car_data[mileage_modified_car_data['EngineV']<enginev_outlier]#0.6 to 6.5
year_modified_car_data=enginev_modified_car_data[enginev_modified_car_data['Year']>year_outlier]

year_modified_car_data.describe(include='all')
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.distplot(car_data['Mileage'])
plt.title('Before Removing Mileage Outliers')

plt.subplot(1,2,2)
sns.distplot(mileage_modified_car_data['Mileage'])
plt.xlim(0,433)
plt.title('After Removing Mileage outliers')
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.distplot(car_data['EngineV'])
plt.title('Before Removing EngineV Outliers')

plt.subplot(1,2,2)
sns.distplot(enginev_modified_car_data['EngineV'])
plt.xlim(0,6)
plt.title('After Removing EngineV Outliers')
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.distplot(car_data['Year'])
plt.title('Before Removing Year Outliers')

plt.subplot(1,2,2)
sns.distplot(year_modified_car_data['Year'])
plt.xlim(1988,2016)
plt.title('After Removing Year Outliers')
```

```python
clean_car_data=year_modified_car_data
clean_car_data
```
